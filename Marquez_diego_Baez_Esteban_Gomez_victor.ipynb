## **Explicaci√≥n del C√≥digo de Ejemplo**

Este c√≥digo es un ejemplo de bloques l√≥gicos anidados que funcionan como un filtro.

El objetivo es crear una lista (resultado) solo con las parejas de n√∫meros (x, y) que cumplan una condici√≥n espec√≠fica.

Para lograrlo, el c√≥digo revisa sistem√°ticamente todas las combinaciones posibles: primero toma un n√∫mero x (del 0 al 9) y luego, para ese x, lo combina con cada n√∫mero y (del 0 al 4).

A cada pareja le aplica una regla: si la multiplicaci√≥n x * y es mayor o igual a 10, la pareja pasa el filtro y se guarda en la lista final. Al final, se imprime la lista con todas las parejas que cumplieron la condici√≥n.

```python
resultado = []
for x in range(10):
    for y in range(5):
        if x * y >= 10:
            resultado.append((x, y))
print(resultado)
```

### **Statsmodels üìà**
<p align="center">
<img src="https://avatars.githubusercontent.com/u/4971254?s=280&v=4" alt="Logo de Statsmodels" width="200">
</p>

Es un laboratorio de estad√≠stica. Se usa para aplicar modelos estad√≠sticos m√°s avanzados, como las regresiones lineales, y realizar pruebas de hip√≥tesis.
P√°gina oficial: www.statsmodels.org

```python
import statsmodels.api as sm
import numpy as np

# Datos: Gasto en publicidad (X) y Ventas (y)
gasto_publicidad = np.array([100, 150, 200, 250, 300])
ventas = np.array([500, 600, 650, 780, 850])

# Preparar los datos para el modelo
X = sm.add_constant(gasto_publicidad)

# Crear y ajustar el modelo de regresi√≥n lineal
modelo = sm.OLS(ventas, X).fit()

# Imprimir el resumen con los resultados estad√≠sticos
print(modelo.summary())
```

## **Notaci√≥n Big-O y orden de rapidez**

La notaci√≥n Big-O describe c√≥mo crece el tiempo de ejecuci√≥n (o uso de memoria) de un algoritmo cuando aumenta el tama√±o de entrada
ùëõ
n.
En general, mientras m√°s lento crece la funci√≥n, m√°s r√°pido ser√° el algoritmo para entradas grandes.

Orden com√∫n de complejidades (de m√°s r√°pida a m√°s lenta):

O(1) ‚Äì Tiempo constante: no importa el tama√±o de entrada (ej. acceso a un elemento en una lista).

O(log n) ‚Äì Logar√≠tmica: crece muy lentamente (ej. b√∫squeda binaria).

O(n) ‚Äì Lineal: el tiempo crece proporcionalmente al tama√±o (ej. recorrer una lista).

O(n log n) ‚Äì Lineal-logar√≠tmica: t√≠pico de algoritmos de ordenamiento eficientes (ej. mergesort).

O(n¬≤) ‚Äì Cuadr√°tica: se vuelve lenta con listas grandes (ej. burbuja, inserci√≥n simple).

O(2‚Åø) ‚Äì Exponencial: se vuelve impr√°ctica muy r√°pido (ej. fuerza bruta en problemas combinatorios).

O(n!) ‚Äì Factorial: extremadamente lenta incluso para valores peque√±os (ej. permutaciones completas).

```python
# Big-O demo con 7 complejidades y animaci√≥n simple en Colab/Jupyter
# Reglas cumplidas: matplotlib, una sola gr√°fica, sin estilos/colores espec√≠ficos.

import time
import math
import numpy as np
import matplotlib.pyplot as plt
from IPython.display import clear_output, display

# --------- Funciones representativas ---------
def op_constante(n):
    # O(1): hace trabajo constante, independiente de n
    a = 42
    a += 1
    a *= 2
    return a

def op_log(n):
    # O(log n): duplica hasta pasar n
    x = 1
    while x < n:
        x *= 2

def op_lineal(n):
    # O(n): recorre n pasos
    for _ in range(n):
        pass

def op_nlogn(n):
    # O(n log n): n veces un trabajo logar√≠tmico
    for _ in range(n):
        x = 1
        while x < n:
            x *= 2

def op_cuadratica(n):
    # O(n^2): dos bucles anidados
    for _ in range(n):
        for _ in range(n):
            pass

def op_exponencial(n):
    # O(2^n): simular 2**n "unidades de trabajo"
    # n peque√±os/medianos para no bloquear
    for _ in range(2 ** n):
        pass

def op_factorial(n):
    # O(n!): simular n! "unidades de trabajo" (n muy peque√±o)
    total = math.factorial(n)
    for _ in range(total):
        pass

# --------- Configuraci√≥n de tama√±os ---------
# Elegidos para que la animaci√≥n sea perceptible pero breve
experimentos = [
    ("O(1)", op_constante, [10, 10**3, 10**6, 10**9]),
    ("O(log n)", op_log, [10**3, 10**4, 10**5, 10**6, 10**7]),
    ("O(n)", op_lineal, [2_000, 8_000, 20_000, 40_000, 80_000]),
    ("O(n log n)", op_nlogn, [300, 800, 1_600, 3_200, 6_400]),
    ("O(n^2)", op_cuadratica, [60, 100, 140, 180, 220]),
    ("O(2^n)", op_exponencial, [12, 14, 16, 18, 20, 22]),     # 2^22 ‚âà 4.2M
    ("O(n!)", op_factorial, [7, 8, 9, 10])                    # 10! ‚âà 3.6M
]

# --------- Estructuras para resultados ---------
resultados = {nombre: {"n": [], "t": []} for nombre, _, _ in experimentos}

# --------- Crear la figura (una sola) ---------
fig, ax = plt.subplots(figsize=(8, 5))
ax.set_title("Crecimiento temporal (Big-O) ‚Äî 7 complejidades (animado)")
ax.set_xlabel("Tama√±o de entrada n")
ax.set_ylabel("Tiempo (segundos)")
ax.set_xscale("log")
ax.set_yscale("log")
ax.grid(True)
plt.tight_layout()

# --------- Animaci√≥n paso a paso ---------
for nombre, fn, ns in experimentos:
    for n in ns:
        t0 = time.perf_counter()
        fn(n)
        t1 = time.perf_counter()
        dt = t1 - t0

        resultados[nombre]["n"].append(n)
        resultados[nombre]["t"].append(dt)

        # Redibujar
        clear_output(wait=True)
        ax.cla()
        ax.set_title("Crecimiento temporal (Big-O) ‚Äî 7 complejidades (animado)")
        ax.set_xlabel("Tama√±o de entrada n")
        ax.set_ylabel("Tiempo (segundos)")
        ax.set_xscale("log")
        ax.set_yscale("log")
        ax.grid(True)

        # Dibujar cada serie disponible
        for serie_nombre, data in resultados.items():
            if data["n"]:
                ax.plot(data["n"], data["t"], marker="o", label=serie_nombre)

        ax.legend(loc="best")
        ax.set_ylim(bottom=1e-6)

        display(fig)
        print(f"Registrado {nombre} con n={n} ‚Üí {dt:.6f} s")

        # Pausa corta para visualizar la progresi√≥n
        time.sleep(0.35)

# Nota: si alguna serie tarda demasiado en tu equipo, reduce los tama√±os de 'n' correspondientes.

```

### **Construir un data frame (CSV) de ejemplo**

En esta secci√≥n, vamos a "inventar" un conjunto de datos para la clase usando las librer√≠as numpy y pandas. A esta tabla la llamaremos da.

El c√≥digo crea una tabla (DataFrame) con 250 filas y le a√±ade columnas con datos generados aleatoriamente que simulan una encuesta de salud. Adem√°s, calcula el √çndice de Masa Corporal (IMC) y agrega intencionalmente algunos valores vac√≠os para practicar c√≥mo se manejan los datos incompletos en un escenario real.

Cada variable (columna) representa lo siguiente:

*  SEQN: Un identificador √∫nico para cada persona en el estudio.

*  RIAGENDR: G√©nero, representado con n√∫meros (1 y 2). Es una variable categ√≥rica.

*  RIDAGEYR: Edad de la persona en a√±os.

*  BMXWT: Peso en kilogramos.

*  BMXHT: Altura en cent√≠metros.

*  BMXBMI: √çndice de Masa Corporal (IMC), que se calcula a partir del peso y la altura.

*  PAIS: Pa√≠s de origen de la persona. Esta es otra variable categ√≥rica que hemos a√±adido.

```python
#Cargada una vez la libreria, no se debe volver a cargar de nuevo.
import numpy as np
import pandas as pd

# Fijar una semilla para que los resultados aleatorios sean siempre los mismos
np.random.seed(42)

# N√∫mero de filas que tendr√° nuestra tabla
n = 250

# Crear el DataFrame
da = pd.DataFrame({
    "SEQN": np.arange(1, n+1),
    "RIAGENDR": np.random.choice([1, 2], size=n, p=[0.5, 0.5]), # 1: Masculino, 2: Femenino
    "RIDAGEYR": np.random.randint(18, 80, size=n),
    "BMXWT": np.round(np.random.normal(70, 12, size=n), 1), # Peso en kg
    "BMXHT": np.round(np.random.normal(168, 10, size=n), 1), # Altura en cm
    "PAIS": np.random.choice(["Colombia", "Mexico", "Argentina"], size=n, p=[0.6, 0.2, 0.2]) # Variable categ√≥rica de texto
})

# Calcular el IMC y a√±adirlo como una nueva columna
da["BMXBMI"] = np.round(da["BMXWT"] / (da["BMXHT"]/100)**2, 1)

# Simular datos faltantes: seleccionar 10 filas al azar y poner NaN en la columna de peso
da.loc[np.random.choice(da.index, size=10, replace=False), "BMXWT"] = np.nan

# Mostrar las primeras 5 filas de nuestra tabla para revisarla
da.head()
```

### **Slicing: Seleccionar Filas y Columnas**

"Slicing" (o rebanar) es la t√©cnica que usamos para seleccionar pedazos espec√≠ficos de nuestra tabla. Podemos querer ver solo una columna (como la edad de todos los participantes) o una fila completa (todos los datos de una persona espec√≠fica).

Pandas nos ofrece varias formas de hacerlo, siendo las m√°s comunes .loc (para seleccionar por nombre) e .iloc (para seleccionar por n√∫mero de posici√≥n).

* **Seleccionar una Columna**
Hay m√∫ltiples maneras de seleccionar una sola columna. Aunque la sintaxis var√≠a, todas las opciones que se muestran a continuaci√≥n logran exactamente lo mismo: aislar la columna RIDAGEYR (edad).

* **Seleccionar una Fila**
Para seleccionar una fila por su posici√≥n, usamos .iloc. Es importante recordar que en Python se empieza a contar desde 0, por lo que .iloc[3] en realidad selecciona la cuarta fila de la tabla.

```python
# --- Formas de seleccionar la columna de edad (RIDAGEYR) ---

# Opci√≥n 1: Usando corchetes y el nombre de la columna (la m√°s com√∫n)
w = da["RIDAGEYR"]

# Opci√≥n 2: Usando .loc para seleccionar por nombre
x = da.loc[:, "RIDAGEYR"] # El ":" significa "todas las filas"

# Opci√≥n 3: Usando un punto (solo funciona si el nombre no tiene espacios)
y = da.RIDAGEYR

# Opci√≥n 4: Usando .iloc para seleccionar por n√∫mero de columna
z = da.iloc[:, da.columns.get_loc("RIDAGEYR")]

# Verificamos que todas las formas dieron el mismo resultado
print("Max RIDAGEYR (cuatro formas):", w.max(), x.max(), y.max(), z.max())


# --- Seleccionar la cuarta fila de la tabla (√≠ndice 3) ---
fila4 = da.iloc[3, :] # El ":" al final significa "todas las columnas"
display(fila4)
```

### **Manejo de Datos Faltantes (NaN)**

En los conjuntos de datos del mundo real, es muy com√∫n que falte informaci√≥n. Pandas representa estos vac√≠os con un valor especial llamado NaN (Not a Number). Antes de hacer cualquier an√°lisis, es crucial identificar cu√°ntos datos faltantes tenemos y decidir qu√© hacer con ellos.


**¬øQu√© hacer con los datos faltantes?**
Una vez identificados, existen varias estrategias comunes para manejarlos. La elecci√≥n depende del problema y de la cantidad de datos que falten.

**Eliminar las filas:** Si tienes muy pocos datos faltantes en un conjunto de datos muy grande, la soluci√≥n m√°s simple es eliminar las filas completas que contienen un NaN.

  Ventaja: Es r√°pido y f√°cil.

  Desventaja: Pierdes la informaci√≥n de las otras columnas de esa misma fila.

**Rellenar (imputar) los valores:** Consiste en reemplazar el NaN con un valor calculado a partir del resto de la columna.

* **Opci√≥n A (para n√∫meros):** Rellenar con el promedio (media) o la mediana de la columna. La mediana suele ser mejor si hay valores at√≠picos.

* **Opci√≥n B (para categor√≠as):** Rellenar con la moda (el valor que m√°s se repite).

  Ventaja: No pierdes filas enteras de datos.

  Desventaja: Puede introducir un peque√±o sesgo en los datos.

**Dejarlos como est√°n**: Algunos modelos de machine learning son capaces de manejar los valores faltantes por s√≠ mismos, aunque no es lo m√°s com√∫n.




El siguiente c√≥digo cuenta cu√°ntos valores faltantes (NaN) y no faltantes hay en la columna de peso (BMXWT).

```python
faltantes_bmxwt = pd.isnull(da["BMXWT"]).sum()
no_faltantes_bmxwt = pd.notnull(da["BMXWT"]).sum()

print("BMXWT ‚Üí faltantes:", faltantes_bmxwt, "| no faltantes:", no_faltantes_bmxwt)
```

### **Visualizar una Variable Num√©rica (Histograma)**

Para las variables num√©ricas como la edad (RIDAGEYR), el histograma es el gr√°fico ideal. Agrupa los n√∫meros en "cajones" o intervalos llamados bins y nos muestra cu√°ntos datos caen en cada uno.

Para decidir cu√°ntos bins usar, una regla cl√°sica es la Regla de Sturges, que busca un balance para que el gr√°fico sea legible. La f√≥rmula es k = 1 + 3.322 * log10(n), donde n es el n√∫mero de datos. Existen otras reglas como la de Scott o Freedman-Diaconis que tambi√©n son muy √∫tiles.

```python
# --- Histograma para la Edad (RIDAGEYR) ---

# 1. Calcular el n√∫mero de bins usando la Regla de Sturges
n = len(da)
k = int(1 + 3.322 * np.log10(n))
print(f"N√∫mero de datos (n): {n}")
print(f"Bins sugeridos por Sturges (k): {k}")


# 2. Crear el histograma
plt.figure(figsize=(10, 5))
plt.hist(da["RIDAGEYR"].dropna(), bins=k, color='skyblue', edgecolor='black')

# A√±adir t√≠tulos y etiquetas
plt.title('Distribuci√≥n de Edades (RIDAGEYR)')
plt.xlabel('Edad (a√±os)')
plt.ylabel('Frecuencia (cantidad de personas)')
plt.grid(axis='y', alpha=0.75)

# Mostrar el gr√°fico
plt.show()
```

### **Visualizar una Variable Categ√≥rica (Gr√°fico de Barras)**
Para variables categ√≥ricas como el pa√≠s (PAIS), donde los datos son etiquetas de texto, el gr√°fico de barras es perfecto. Simplemente cuenta cu√°ntas veces aparece cada categor√≠a y muestra el total con una barra.

```python
# --- Gr√°fico de Barras para el Pa√≠s (PAIS) ---

# 1. Contar la frecuencia de cada categor√≠a
conteo_paises = da['PAIS'].value_counts()
print("Conteo por pa√≠s:")
print(conteo_paises)


# 2. Crear el gr√°fico de barras
plt.figure(figsize=(8, 5))
plt.bar(conteo_paises.index, conteo_paises.values, color=['#0033A0', '#FCD116', '#CE1126'])

# A√±adir t√≠tulos y etiquetas
plt.title('Distribuci√≥n de Personas por Pa√≠s')
plt.xlabel('Pa√≠s')
plt.ylabel('Cantidad de Personas')

# Mostrar el gr√°fico
plt.show()
```
